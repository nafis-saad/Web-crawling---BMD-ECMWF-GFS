```python
url = "http://bmd.gov.bd/p/Weather-Forecast/"
page = requests.get(url, headers={"User-Agent": "XY"})
page.status_code

print(page.text)
```

```python
!pip install beautifulsoup4
from bs4 import BeautifulSoup
soup = BeautifulSoup(page.content, "html.parser")
links = [ ]
for link in soup.findAll("a"):
    links.append(link.get("href"))
print(links)
```

```python
import time
today = time.strftime("%Y/%m/%d")
print(today)


import re
import os
import datetime
for element in links:
  z = re.match(f"^http://bmd\\.gov\\.bd/file/{today}/pdf/.*\\.pdf$", element)
  if z:
      break

type(z)
pdf = str(z.group())

output_dir = '/content/drive/MyDrive/Python_Practice'

response = requests.get(pdf, headers={"User-Agent": "XY"})

if response.status_code == 200:
  file_path = os.path.join(output_dir, os.path.basename(pdf))
  with open(file_path, "wb") as f:
    f.write(response.content)




```
